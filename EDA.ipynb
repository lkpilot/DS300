```python
df_job = pd.read_csv(PATH + "/Job_data_final.csv")
df_job.head(10)
```

## Các từ xuất hiện trong cột Industry của job dataset

```python
d_frame = df_job['Industry'][:2000]

plt.subplots(figsize=(10,10))
wordcloud = WordCloud(
                   background_color='White',
                        width = 700,
                        height = 400
    ).generate(" ".join(d_frame))

plt.imshow(wordcloud)
plt.axis('off')
```

## Các công việc phổ biến nhất

## Mức kinh nghiệm phổ biến của ứng viên và số năm kinh nghiệm của công việc

```python
df_c=df_user
plt.figure(figsize=(7,7))

sns.color_palette("magma", as_cmap=True)
ax = sns.countplot(y="Work Experience", data=df_c,palette="Greens").set_title('Mức kinh nghiệm của ứng cử viên', fontsize=25)
```

```python
df_c=df_job
plt.figure(figsize=(7,7))
sns.set(style="darkgrid")
ax = sns.countplot(y="Years of Experience", data=df_c,palette="Greens").set_title('Mức kinh nghiệm mà nhà tuyển dụng để ra', fontsize=25)
```

## Mức lương mong muốn của ứng viên và mức lương đề xuất của doanh nghiệp

```python
df_c=df_job[df_job['Salary'] != 'Thỏa thuận']
plt.figure(figsize=(7,7))
sns.set(style="darkgrid")
ax = sns.countplot(y="Salary", data=df_c).set_title('Mức lương để ra của nhà tuyển dụng', fontsize=25)
```

```python
df_c=df_user[df_user['Desired Salary'] != 'Thỏa thuận']
plt.figure(figsize=(7,7))
sns.set(style="darkgrid")
ax = sns.countplot(y="Desired Salary", data=df_c).set_title('Mức lương mong muốn của ứng cử viên',fontsize=25)
```

## Độ tuổi trung bình của ứng viên

```python
plt.figure(dpi=125)
sns.distplot(a=df_user['Age'],kde=False,bins=20,color='red')
plt.axvline(x=np.mean(df_user['Age']),c='green',ls='--',label='Mean Age of All candidates')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```
